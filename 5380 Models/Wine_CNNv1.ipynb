```python
winedata = pd.read_csv(r"winedata.csv")
wine_df = pd.DataFrame(winedata)
wine_df
```

```python
train_x = wine_df
train_x = train_x.drop("Class Type", axis = 1)
train_x
```

```python
wine_df.shape
```

```python
x = wine_df.values #features
x
```

```python
y = wine_df["Class Type"]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_x, y, train_size = 0.80)
```

```python
print("shape of x_train = ", x_train.shape)
print("shape of y_train = ", y_train.shape)
print("shape of x_test = ", x_test.shape)
print("shape of y_test = ", y_test.shape)
```

```python
x_train.shape
```

```python
model = Sequential()
model.add(Conv2D(len(x_train), (2,2), activation = 'relu'))
model.add(Dropout(0.05))
model.add(Flatten())
model.add(Dense(128, activation = 'relu'))
model.add(Conv2D(3, (2,2), activation = 'relu'))
  # Compile model
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

accuracy = model.evaluate(x_test, y_test)
print(accuracy)
```
